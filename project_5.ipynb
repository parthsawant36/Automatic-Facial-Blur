```python
import cv2 as cv 
import time
import numpy as np
import time
```

```python
img = cv.imread("./images/happy_parents.jpg")
cv.imshow("Og" , img)
cv.waitKey(10000)
cv.destroyAllWindows()
```

```python
face_detection_model = cv.dnn.readNetFromCaffe('./model/deploy.prototxt.txt' , './model/res10_300x300_ssd_iter_140000_fp16.caffemodel')
```

```python
def face_blur(img) :
    image = img.copy()
    flag = 0
    face_mask = np.zeros(image.shape[:2] , dtype = "uint8")
    
    blob = cv.dnn.blobFromImage(image, 1 , (300,300) , (104,177,123) ,swapRB = True)
    face_detection_model.setInput(blob)
    
    detections = face_detection_model.forward()
    
    
    h,w = image.shape[0] , image.shape[1]
    for i in range (detections.shape[2]) :
        confidence = detections[0,0,i,2]
        if confidence > 0.5:
            box = detections[0,0,i,3:7]*np.array([w,h,w,h])
            box = box.astype(int)
            cv.rectangle(face_mask, (box[0] ,box[1]) , (box[2] , box[3]) , 255 , -1)
            flag += 1
            print(confidence)
    
    #apply mask image to original image
    inv_mask = cv.bitwise_not(face_mask)
    #(face visible , rest black)
    bg_img = cv.bitwise_and(image,image,mask = face_mask)
    #(rest visible , face black)
    fg_img = cv.bitwise_and(image,image,mask = inv_mask)
    #blur the face , using bg_img as only that has the face 
    bg_img_blur = cv.GaussianBlur(bg_img , (51,51),0)
    
    #use bitwise and
    face_blur = cv.add(fg_img , bg_img_blur)
    
    return face_blur
```

```python
my_blur = face_blur(img)

cv.imshow("Og img" , img)
cv.imshow("Fcae Blur Image" , my_blur)
cv.waitKey(10000)
cv.destroyAllWindows()
```

```python
cap = cv.VideoCapture(0)

while True :
    start_time = time.time()
    ret , frame = cap.read()

    if ret == False :
        break

    
    if cv.waitKey(33) == 27 :
        break

    time_taken = time.time() - start_time
    fps = 1 / time_taken
    cv.putText(frame, "FPS = {:.0f}".format(fps), (30, 40),
               cv.FONT_HERSHEY_SIMPLEX, 2, (255,255, 255), 6)
    blur_img = face_blur(frame)

    cv.imshow("og" , frame)
    cv.imshow("Real Time Face Blur" , blur_img)

    

cap.release()
cv.destroyAllWindows()
```
